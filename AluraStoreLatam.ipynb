### Importaci√≥n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An√°lisis de facturaci√≥n



```python
# Normalizar nombres de columnas para evitar errores
def limpiar_columnas(df):
    df.columns = df.columns.str.strip().str.lower()  # elimina espacios y pone en min√∫sculas
    return df

# Aplicar a cada tienda
tienda1 = limpiar_columnas(tienda1)
tienda2 = limpiar_columnas(tienda2)
tienda3 = limpiar_columnas(tienda3)
tienda4 = limpiar_columnas(tienda4)

# Ahora todas las columnas est√°n en min√∫sculas y sin espacios al principio/final
ingresos_t1 = tienda1['precio'].sum()
ingresos_t2 = tienda2['precio'].sum()
ingresos_t3 = tienda3['precio'].sum()
ingresos_t4 = tienda4['precio'].sum()

# Mostrar resultados
print("\nüí∞ Ingresos Totales por Tienda:")
print(f"Tienda 1: ${ingresos_t1:,.2f}")
print(f"Tienda 2: ${ingresos_t2:,.2f}")
print(f"Tienda 3: ${ingresos_t3:,.2f}")
print(f"Tienda 4: ${ingresos_t4:,.2f}")

```

# 2. Ventas por categor√≠a

```python
import pandas as pd

# Cargar los datos desde las URLs
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Revisar columnas disponibles
print(tienda1.columns)

# Agrupar por categor√≠a y contar productos vendidos
ventas_cat_t1 = tienda1['Categor√≠a del Producto'].value_counts().reset_index()
ventas_cat_t1.columns = ['Categor√≠a del Producto', 'Ventas']
ventas_cat_t2 = tienda2['Categor√≠a del Producto'].value_counts().reset_index()
ventas_cat_t2.columns = ['Categor√≠a del Producto', 'Ventas']
ventas_cat_t3 = tienda3['Categor√≠a del Producto'].value_counts().reset_index()
ventas_cat_t3.columns = ['Categor√≠a del Producto', 'Ventas']
ventas_cat_t4 = tienda4['Categor√≠a del Producto'].value_counts().reset_index()
ventas_cat_t4.columns = ['Categor√≠a del Producto', 'Ventas']

# Mostrar las categor√≠as m√°s vendidas de cada tienda
print("\nüì¶ Categor√≠as m√°s vendidas - Tienda 1")
print(ventas_cat_t1)

print("\nüì¶ Categor√≠as m√°s vendidas - Tienda 2")
print(ventas_cat_t2)

print("\nüì¶ Categor√≠as m√°s vendidas - Tienda 3")
print(ventas_cat_t3)

print("\nüì¶ Categor√≠as m√°s vendidas - Tienda 4")
print(ventas_cat_t4)
```

# 3. Calificaci√≥n promedio de la tienda


```python
# Verificar las columnas para encontrar la de calificaci√≥n
print(tienda.columns)

# Asumimos que la columna de calificaci√≥n se llama 'Calificaci√≥n del Producto'
# Si el nombre es diferente, lo corregimos aqu√≠

# Calcular la calificaci√≥n promedio para cada tienda
calificacion_tienda1 = tienda['Calificaci√≥n'].mean()
calificacion_tienda2 = tienda2['Calificaci√≥n'].mean()
calificacion_tienda3 = tienda3['Calificaci√≥n'].mean()
calificacion_tienda4 = tienda4['Calificaci√≥n'].mean()

# Mostrar los resultados
print("\nüìä Calificaci√≥n promedio por tienda:")
print(f"Tienda 1: {calificacion_tienda1:.2f}")
print(f"Tienda 2: {calificacion_tienda2:.2f}")
print(f"Tienda 3: {calificacion_tienda3:.2f}")
print(f"Tienda 4: {calificacion_tienda4:.2f}")

```

# 4. Productos m√°s y menos vendidos

```python
import matplotlib.pyplot as plt

# Agrupar por producto y contar las ventas de cada uno
ventas_producto_t1 = tienda['Producto'].value_counts().reset_index()
ventas_producto_t1.columns = ['Producto', 'Ventas']

ventas_producto_t2 = tienda2['Producto'].value_counts().reset_index()
ventas_producto_t2.columns = ['Producto', 'Ventas']

ventas_producto_t3 = tienda3['Producto'].value_counts().reset_index()
ventas_producto_t3.columns = ['Producto', 'Ventas']

ventas_producto_t4 = tienda4['Producto'].value_counts().reset_index()
ventas_producto_t4.columns = ['Producto', 'Ventas']

# Mostrar los 5 productos m√°s vendidos y los menos vendidos
print("\nüì¶ Productos m√°s vendidos - Tienda 1")
print(ventas_producto_t1.head())

print("\nüì¶ Productos menos vendidos - Tienda 1")
print(ventas_producto_t1.tail())

# Similar para las otras tiendas
print("\nüì¶ Productos m√°s vendidos - Tienda 2")
print(ventas_producto_t2.head())

print("\nüì¶ Productos menos vendidos - Tienda 2")
print(ventas_producto_t2.tail())

print("\nüì¶ Productos m√°s vendidos - Tienda 3")
print(ventas_producto_t3.head())

print("\nüì¶ Productos menos vendidos - Tienda 3")
print(ventas_producto_t3.tail())

print("\nüì¶ Productos m√°s vendidos - Tienda 4")
print(ventas_producto_t4.head())

print("\nüì¶ Productos menos vendidos - Tienda 4")
print(ventas_producto_t4.tail())

# Visualizar los 5 productos m√°s vendidos en un gr√°fico de barras
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.flatten()

# Gr√°ficos de barras para cada tienda
for i, (df, nombre) in enumerate(zip([ventas_producto_t1, ventas_producto_t2, ventas_producto_t3, ventas_producto_t4],
                                    ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])):
    axes[i].bar(df['Producto'].head(10), df['Ventas'].head(10), color='lightblue')
    axes[i].set_title(f'{nombre} - Top 10 Productos M√°s Vendidos')
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].set_ylabel('Cantidad de Ventas')
    axes[i].set_xlabel('Producto')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Funci√≥n para graficar dispersi√≥n de coordenadas
def graficar_lat_lon(df, nombre_tienda):
    plt.figure(figsize=(8, 6))
    plt.scatter(df['lon'], df['lat'], alpha=0.5, c='blue', edgecolors='k')
    plt.title(f'Distribuci√≥n Geogr√°fica de Ventas - {nombre_tienda}')
    plt.xlabel('Longitud')
    plt.ylabel('Latitud')
    plt.grid(True)
    plt.show()

# Graficar para cada tienda
graficar_lat_lon(tienda, 'Tienda 1')
graficar_lat_lon(tienda2, 'Tienda 2')
graficar_lat_lon(tienda3, 'Tienda 3')
graficar_lat_lon(tienda4, 'Tienda 4')
```
